# Matplotlib Filling Area on Line Plot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data_fill.csv')
df
```

```python
ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

Fill area below python salary

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# fill
plt.fill_between(ages, py_salaries)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# fill
plt.fill_between(ages, py_salaries, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# fill
overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries <= overall_median), color='red', interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```
