# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
slices = [60,40]
plt.pie(slices)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Add label

```python
label = ['sixty','fourty']
plt.pie(slices, labels=label)


plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

add edge color

```python
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

The value shouldn't 100 in total

```python
slices = [60, 40, 30, 10]
labels = ['Sixty', 'Forty', 'Thirty', 'Ten']
#colors = ['blue', 'red', 'yellow', 'green']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor': 'black'})

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Use data of Number of People surveyed by StackOverflow

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 
          23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]

labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java',
          'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript',
          'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

It is not good to use pie chart for data that has more than 5 items

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Ephasize an items

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Add shadow

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Start at certain point

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, startangle=90)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

add percentage

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, startangle=90,
       autopct = '%1.1f%%')

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```
